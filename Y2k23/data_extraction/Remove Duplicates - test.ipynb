```python
import os
import json
import pandas as pd
```

```python
test_filepath = "../data/test.json"
```

```python
df_cs = pd.DataFrame()
count = 0
dct_events = {}
for each_event in open(test_filepath, 'r'):
    count += 1
    print(count)
    if not dct_events.get(each_event):
            dct_events[each_event] = 0
    dct_events[each_event] += 1
```

```python
df_cs = pd.DataFrame()
for each_key in dct_events.keys():
    event = eval(each_key)
    if event.get("event"):
        event = event.get("event")
    #event = event.get("event") or event.get("AlertID")
    print(event)
    df_cs = df_cs.append(event, ignore_index=True)
```
